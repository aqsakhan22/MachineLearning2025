```python
x=df.drop(columns=['Salary']) # drop salary ka kiya baki jo hai wo ayga like yearsExperience
#x is independet
# salary jo hai y may assign kiya means it is independent variable
y=df['Salary'] # assignment

y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(x,y,test_size=0.2,random_state=0)
# we set random_state = 0 for not show random data
```

```python
# X_train
X_train
# len(X_train)
```

```python
from sklearn.linear_model import LinearRegression
# tarin test k baad select kerna huga k knsa model use kerna hai
# like hum kis mmodel wo data dygy
```

```python
model = LinearRegression()
```

```python
# Training ndata 
model.fit(X_train,Y_train) # prroviding data to model for learn data 
```

```python
# now test that data 
y_pred=model.predict(X_test)
# x_test may year hai is saal k experience ki kitni salary hugi 
```

```python
pd.DataFrame({'y_test':Y_test,'y_pred':y_pred,'MAE':Y_test - y_pred})
# actual may or predict may kitna difference hai
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
import numpy as np
#
print('Mean square error',mean_squared_error(Y_test,y_pred)) # ismay data bara ata haia tbhi hum rmse lagaty hai no bara hai means sahi nh
print('Mean absolute error',mean_absolute_error(Y_test,y_pred))
print('Root Mean square error',np.sqrt(mean_squared_error(Y_test,y_pred))) # rmse jitna chota hai utna acha huta hai 
print('R2-Score',r2_score(Y_test,y_pred))
# r2 score jitna 1 k kareeb huta hai utna acha huta hai
# error jub bht bara dikhara hu tou tbhi hum rmse use ker rahy hai
```
